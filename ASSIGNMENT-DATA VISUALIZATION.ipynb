## ASSIGNMENT-DATA VISUALISATION

## Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## importing dataset

```python
df=sns.load_dataset("planets")
df.head(10)
```

## lineplot

```python
sns.lineplot(x="orbital_period",y="mass",data=df)
```

```python
sns.lineplot(x="orbital_period",y="distance",data=df)
```

```python
sns.lineplot(x="orbital_period",y="distance",data=df,hue="mass",color="blue",linewidth=2)
```

## distplot

```python
sns.distplot(df["orbital_period"])
```

```python
sns.distplot(df["distance"],kde=False,bins=50)
```

## histplot

```python
df.hist(figsize=(30,25),bins=40)
```

## barplot

```python
sns.barplot(x="orbital_period",y="mass",data=df,estimator=np.mean)
```

```python
sns.barplot(x="mass",y="distance",data=df,estimator=np.std)
```

## jointplot

```python
sns.jointplot(x="orbital_period",y="mass",data=df,kind="scatter")
```

## pairplot

```python
sns.pairplot(df)
```

## rugplot and kdeplot

```python
sns.rugplot(df["orbital_period"])
sns.kdeplot(df["orbital_period"])
```

## boxplot

```python
sns.boxplot(x="orbital_period",y="mass",data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,palette="rainbow",orient="h")
```

## voilinplot

```python
sns.violinplot(x="distance",y="mass",data=df,palette="rainbow")
```

```python
sns.violinplot(x="distance",y="orbital_period",data=df,palette="Set1")
```

## stripplot

```python
sns.stripplot(x="distance",y="mass",data=df)
```

## swarmplot

```python
sns.swarmplot(x="distance",y="mass",data=df,palette="Set1",split=True)
```

## heatmap

```python
sns.heatmap(df.corr())
```

```python
v=df.pivot_table(values="orbital_period",index="mass",columns="distance")
sns.heatmap(v)
```

## catplot

```python
sns.catplot(x="distance",y="mass",data=df,kind="boxen")
```

## lmplot

```python
sns.lmplot(x="mass",y="orbital_period",height=3,aspect=6,data=df)
```
