# MACHINE LEARNING MODELS:-

```python
X=data[data.columns[0:13]]
y=data["target"]
```

```python
#NORMALIZE DATA
X_norm= (X - np.min(X))/(np.max(X)- np.min(X)).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X_norm,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression(random_state=10)
```

```python
lg.fit(X_train,y_train)
```

```python
accuracy_scores =[]
accuracy=round(lg.score(X_test,y_test)*100,2)
accuracy_scores.append(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=lg.predict(X_test)
log_con=pd.DataFrame(confusion_matrix(y_test,y_pred))
log_con.columns=["PREDICTED: NO DISEASE","PREDICTED:DISEASE"]
log_con.rename(index={0:"ACTUAL:NO DISEASE",1:"ACTUAL:DISEASE"},inplace=True)
sns.heatmap(log_con,annot=True,cmap="GnBu")
plt.title("CONFUSION MATRIX FOR LOGISTIC REGRESSION")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
#import matplotlib.pyplot as plt 
```

```python
auc=roc_auc_score(y_test, y_pred)
y_prob=lg.predict_proba(X_test)[:,1]
fpr, tpr, ther=roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label="Logistic Regression(auc=%0.2f)"%auc)
plt.plot([0, 1], [0, 1], "o--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FALSE POSITIVE RATE')
plt.ylabel('TRUE POSITIVE RATE')
plt.title("ROC CURVE")
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators = 1000, random_state = 10, max_depth=10)
```

```python
rf_model=rf.fit(X_train, y_train)
rf_accuracy= round(rf_model.score(X_test, y_test)*100,2)
print("The accuracy of the random forest model is: %.2f percent" %rf_accuracy)
accuracy_scores.append(rf_accuracy)
```

```python
rf_y_predict=rf_model.predict(X_test)
rf_con= pd.DataFrame(confusion_matrix(y_test, rf_y_predict))
rf_con.columns=["PREDICTED: No Disease", "PREDICTED: Disease"]
rf_con.rename(index={0: "ACTUAL: No Disease", 1: "ACTUAL: Disease"},inplace=True)
sns.heatmap(rf_con, annot=True, cmap="GnBu")
plt.title("CONFUSION MATRIX OF RANDOM FOREST ALGORITHM")
plt.show()
```

```python
print(classification_report(y_test,rf_y_predict))
```

```python
rf_auc=roc_auc_score(y_test, rf_y_predict)
rf_y_prob=rf_model.predict_proba(X_test)[:,1]
rf_fpr, rf_tpr, rf_ther=roc_curve(y_test, rf_y_prob)
plt.plot(rf_fpr, rf_tpr, label="Random Forest(auc=%0.2f)"%rf_auc)
plt.plot([0, 1], [0, 1], "o--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FALSE POSITIVE RATE')
plt.ylabel('TRUE POSITIVE RATE')
plt.title("ROC CURVE")
plt.legend(loc="lower right")
plt.show
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_scores=[]
test_scores=[]
for k in range (1,40,2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score=knn.score(X_train, y_train)
    test_score=knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
```

```python
plt.plot(range(1, 40, 2), train_scores, marker='o')
plt.plot(range(1, 40, 2), test_scores, marker="x")
plt.xlabel("K NEIGHBOURS")
plt.ylabel("TEST ACCURACY SCORE")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn_model= knn.fit(X_train, y_train)
```

```python
knn_accuracy= knn.score(X_test, y_test)*100
print("The accuracy of the KNN model is: %.2f percent" %knn_accuracy)
accuracy_scores.append(knn_accuracy)
```

```python
knn_y_predict=knn_model.predict(X_test)
knn_con= pd.DataFrame(confusion_matrix(y_test, knn_y_predict))
knn_con.columns=["PREDICTED: No Disease", "PREDICTED: Disease"]
knn_con.rename(index={0: "ACTUAL: No Disease", 1: "ACTUAL: Disease"},inplace=True)
sns.heatmap(knn_con, annot=True, cmap="GnBu")
plt.title("CONFUSION MATRIX FOR K-NEAREST NEIGHBOR ")
plt.show()
```

```python
print(classification_report(y_test, knn_y_predict))
```

```python
knn_auc=roc_auc_score(y_test, knn_y_predict)
knn_y_prob=knn_model.predict_proba(X_test)[:,1]
knn_fpr, knn_tpr, knn_ther=roc_curve(y_test, knn_y_prob)
plt.plot(knn_fpr, knn_tpr, label="KNN(auc=%0.2f)"%knn_auc)
plt.plot([0, 1], [0, 1], "o--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FALSE POSITIVE RATE')
plt.ylabel('TRUE POSITIVE RATE')
plt.title("ROC CURVE")
plt.legend(loc="lower right")
plt.show
```

```python
from sklearn.svm import SVC
```

```python
svm_model=SVC(random_state=10,kernel='linear')
svm_model.fit(X_train,y_train)
```

```python
svm_accuracy= svm_model.score(X_test,y_test)*100
print("The accuracy of the SVM model is: %.2f percent" %svm_accuracy)
accuracy_scores.append(svm_accuracy)
```

```python
svm_y_predict=svm_model.predict(X_test)
svm_con= pd.DataFrame(confusion_matrix(y_test, svm_y_predict))
svm_con.columns=["PREDICTED: No Disease", "PREDICTED: Disease"]
svm_con.rename(index={0: "ACTUAL: No Disease", 1: "ACTUAL: Disease"},inplace=True)
sns.heatmap(svm_con, annot=True, cmap="GnBu")
plt.title("CONFUSION MATRIX FOR SUPPORT VECTOR MACHINE ALGORITHM")
plt.show()
```

```python
print(classification_report(y_test, svm_y_predict))
```

```python
sns.set(style="whitegrid")
colors = ['lightcoral' if (x < max(accuracy_scores)) else 'cadetblue' for x in accuracy_scores ]
sns.barplot(x=[1,2,3,4], y=accuracy_scores,palette=colors)
plt.xticks(np.arange(4),labels=("Logistic Reg", "Random Forest", "KNN", "SVM"))
plt.title("ACCURACY SCORES OF 4 ML MODELS")
```
